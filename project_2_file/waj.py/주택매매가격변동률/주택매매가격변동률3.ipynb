```python
import pandas as pd
```

```python
df = pd.read_csv("201001_202112_월별_시도별_주택매매가격변동률.csv",encoding="cp949")
df
```

```python
df = df.drop([0],axis=0)
df = df.reset_index()
```

```python
df
```

```python
df = df.rename(columns={'행정구역별(1)' : "시도"})
```

```python
df = df.replace('-', '0')
```

```python
df = df.drop(columns=['index'])
```

```python
df = df.T.reset_index()
```

```python
df
```

```python
df = df.rename(columns = df.iloc[0])
```

```python
df = df.drop([0],axis=0)
```

```python
df = df.reset_index()
```

```python
df = df.melt(id_vars=['시도'], value_vars=['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
       '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도',
       '경상북도', '경상남도', '제주특별자치도'], var_name='11', value_name='주택가격변동률')
```

```python
df = df.rename(columns = {'11': '시도', '시도' : '년월'})
```

```python
df
```

```python
df['년월'] = df['년월'].str.replace('.', '_')
```

```python
df
```

```python
df
```

```python
mask = df['년월'].str.endswith('_1')
```

```python
df.loc[mask, '년월'] = df.loc[mask, '년월'].str.replace('_1', '_10')
```

```python
df
```

```python
##년월만들기
df['월'] = df['년월'].str.split('_').str.get(1)
df['연도'] = df['년월'].str.split('_').str.get(0)
```

```python
df.columns
```

```python
df = df[['년월', '연도','월', '시도', '주택가격변동률']]
```

```python
df['월'] = df['월'].astype("int64")
df['연도'] = df['연도'].astype("int64")
```

```python
for i in range(0, 2249, 1):
    if df.loc[i, '월'] == 1:
         df.loc[i, '월'] = '01'
    elif df.loc[i, '월'] == 2:
         df.loc[i, '월'] = '02'
    elif df.loc[i, '월'] == 3:
         df.loc[i, '월'] = '03'
    elif df.loc[i, '월'] == 4:
         df.loc[i, '월'] = '04'
    elif df.loc[i, '월'] == 5:
         df.loc[i, '월'] = '05'
    elif df.loc[i, '월'] == 6:
         df.loc[i, '월'] = '06'
    elif df.loc[i, '월'] == 7:
         df.loc[i, '월'] = '07'
    elif df.loc[i, '월'] == 8:
         df.loc[i, '월'] = '08'
    elif df.loc[i, '월'] == 9:
         df.loc[i, '월'] = '09'
```
