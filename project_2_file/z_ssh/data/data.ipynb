```python
import pandas as pd

def replace_with_row_avg(df, column_name, value):

    row_means = df.mean(axis=1)
    

    df[column_name] = df.apply(lambda row: row[column_name] if row[column_name] != value else row_means[row.name], axis=1)
    
    return df
```
