```python
import pandas as pd
df = pd.read_csv('2010~2022_고용률.csv')
df = df.drop(columns=['Unnamed: 0'])
```

```python
df['연월'] = df['연월'].astype(str)
```

```python
df
```

```python
df['연월'] = df['연월'].str.replace('.', '_')
```

```python
mask = df['연월'].str.endswith('_1')
df.loc[mask, '연월'] = df.loc[mask, '연월'].str.replace('_1', '_10')
```

```python
df = df[['연월', '연도', '월','시도', '고용률']]
```
